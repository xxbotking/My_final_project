```python
"""When people are ordering items online,they find it difficult to know the cost of delivery.
I am creating a model that can accurately predict the cost of delivering an item.
 this is a regression problem.
i am using a dataset from kaggle that contains the details of deliveries in America
The size of the data set was 3000kb
```

```python
##importing tools from data preparation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
##importing dataset 
costdelivery = pd.read_csv(r"C:\Users\XXXBOTKING360\Downloads\archive (1)\CUsersXXXBOTKING360Downloadsarchive (1).csv")
costdelivery
```

```python
##DATA PREPARATION
```

```python
##checking shape of dataset
costdelivery.shape
```

```python
##checking column names
costdelivery.columns
```

```python
##columns to drop
columnstodrop = ['delivery_id', 'delivery_partner', 'delivery_time_hours','expected_time_hours', 'delivery_status']
```

```python
##dropping columns
costdelivery.drop(columns= columnstodrop , inplace=True )
```

```python
##calling cleaned dataset
costdelivery
```

```python
##calling new columns
costdelivery.columns
```

```python
##converting string data to float
dummy=['package_type', 'vehicle_type', 'delivery_mode', 'region',
       'weather_condition', 'distance_km', 'package_weight_kg', 'delayed',
       'delivery_rating', 'delivery_cost']
ctdelivery = pd.get_dummies(costdelivery[dummy], dtype= float)
ctdelivery
```

```python
##checking shape of cleaned dataset
ctdelivery.shape
```

```python
##getting features for model(x_values)
xdelivery = ctdelivery.drop(columns=["delivery_cost"])
```

```python
##calling features(x_values)
xdelivery
```

```python
##getting labels for the model(y_values)
ycost=ctdelivery["delivery_cost"]
```

```python
##calling labels(y_values)
ycost
```

```python
##importing tool to split data for testing and training
from sklearn.model_selection import train_test_split
```

```python
##spliting data
xtrain,xtest,ytrain,ytest = train_test_split(xdelivery,ycost,test_size=0.40)
```

```python
##checking shape of training set and testing set
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
##MODEL SELECTION
```

```python
##importing model/algorithm
from sklearn.ensemble import RandomForestRegressor
```

```python
##setting model parameters
costpredictor = RandomForestRegressor(n_estimators=100)
costpredictor.get_params
```

```python
##MODEL FITTING
```

```python
##fitting model with data
costpredictor.fit(xtrain,ytrain)
```

```python
##EVALUATING MODEL
```

```python
##using model to predict values using xtest
costpredicted=costpredictor.predict(xtest)
```

```python
##checking accuracy of model with the training dataset
cp=costpredictor.score(xtrain,ytrain)
```

```python
##printing accuracy of model
print(f"the accuracy in training is {cp*100:.2f}%")
```

```python
##checking accuracy of model with the testing dataset
cpr=costpredictor.score(xtest,ytest)
```

```python
##showing accuracy
print(f"the accuracy in training is {cpr*100:.2f}%")
```

```python
print(costpredicted)
```

```python
## EXPERIMENTATION
```

```python
##checking accuracy with cross_val_score
import numpy as np
from sklearn.model_selection import cross_val_score
np.random.seed(85)
modelacc= np.mean(cross_val_score(costpredictor,xtest,ytest,cv=2))
print(f"5-fold cross validation score is {modelacc*100:.2f}%")
print("")
```

```python
##getting the best parameters ,accuracy and estimator with gridsearchcv
from sklearn.model_selection import GridSearchCV
np.random.seed(85)
gridparams ={'n_estimators':[i for i in range(100,200,10)]}
gridmodel = GridSearchCV(costpredictor,
                         param_grid=gridparams,
                         cv=3,
                         verbose=1)
gridmodel.fit(xtrain,ytrain)
print(f"the best parameters for the model is {gridmodel.best_params_}")
print(f"with a score of {gridmodel.best_score_*100:.2f}%")
print("")
```

```python
##calling the best model with the best parameters
bestcostpredictor=gridmodel.best_estimator_
bestcostpredictor
```

```python
##fitting the best model with training set
bestcostpredictor.fit(xtrain,ytrain)
```

```python
##checking accuracy of best model with testing set
BCP= bestcostpredictor.score(xtest,ytest)
print(f"the accuracy on a single train/test split is {BCP*100:.2f}%")
```

```python
##tseting accuracy of best model with r2 score
yp= bestcostpredictor.predict(xtest)
from sklearn.metrics import r2_score
r2_score(ytest,yp)
```

```python
##comparing actual data with pedicted data
DF=pd.DataFrame({'ACTUAL COSTS':ytest,'PREDICTED COST':yp})
DF
```
